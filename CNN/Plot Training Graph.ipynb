```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
plt.style.use('ggplot')
```

```python
files = [
#     'dodo2_cnn_rand1_history',
    'dodo_binary_cnn_non_static_history'
]
```

```python
train_acc, train_loss, val_acc, val_loss = [],[],[],[]

for file in files:
    with open( file, 'rb') as input:
        out = pickle.load(input)
    train_acc += out[0]['acc']
    train_loss += out[0]['loss']
    val_acc += out[0]['val_acc']
    val_loss += out[0]['val_loss']
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.rcParams["figure.figsize"] = (10,6)
plt.plot(train_acc)
plt.plot(train_loss)
plt.plot(val_acc)
plt.plot(val_loss)
plt.legend(['Train acc','Train loss','Valid acc', 'Valid loss'], loc=2)
plt.xlabel('Epochs')
plt.ylabel('Loss/Accuracy')

name = 'CNN Binary non-static'
plt.title(name)
plt.savefig( name, dpi= 200, bbox_inches='tight', transparent=False)
plt.show()
```
