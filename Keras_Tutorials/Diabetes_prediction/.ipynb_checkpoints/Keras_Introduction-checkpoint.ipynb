### Loading required libraries

```python
from keras.models import Sequential
from keras.layers import Dense
import matplotlib.pyplot as plt
import numpy
import h5py

# Fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
```

### Loading Dataset

```python
# load pima indians dataset
dataset = numpy.loadtxt("pima-indians-diabetes.csv", delimiter=",")

# split into input (X) and output (Y) variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

### Notes
Fully connected layers $\rightarrow$ Dense  
$1^{st}$ arg $\rightarrow$ **number of neurons**  
$2^{nd}$ arg $\rightarrow$ **Initialization method**  
$3^{rd}$ arg $\rightarrow$ **Activation function**

### Creating model

```python
# create model
model = Sequential()
model.add(Dense(12, input_dim=8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

## 'binary_crossentropy' -> logarithmic loss
## 'adam' -> an efficient gradient descent algorithm
## 'accuracy' -> report classification accuracy as the metric
```

```python
# Get summary of what we have created
model.summary()
```

```python
model.get_config()
# Similarly, can use this output to set a model using: 
# Model.from_config(config) / Sequential.from_config(config)
## model.to_json() and from_json() also available
```

### Training Model 

```python
%%time
# Fit the model
hist = model.fit(X, Y, epochs=150, batch_size=10)

## epochs -> number of epoch
oo## batch_size -> number of observations to be evaluated before updating weights
```

```python
# Visualize output
plt.plot(hist.history['acc'])
plt.plot(hist.history['loss'])
plt.show()
```

```python
# Get the weights and biases of the trained model
wts = model.get_weights()
for i in range(len(wts)):
    print( wts[i].shape)
    
# similarly, we can set wts using: model.set_weights(weights)
# save wts to file: 
# model.save_weights('/home/shikhar/model_wts');
# load wts from file: model.load_weights(filepath, by_name=False)
```

### Predicting on new data

```python
# evaluate the model: Computes the loss on some input data, batch by batch.
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))

# Prediction
predictions = model.predict(X)
rounded = [round(x[0]) for x in predictions]
```
